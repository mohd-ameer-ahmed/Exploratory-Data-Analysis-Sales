```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

```python
data = {
    "Order_ID": [101, 102, 103, 104, 105, 106],
    "Product": ["Laptop", "Mobile", "Tablet", "Laptop", "Mobile", "Tablet"],
    "Category": ["Electronics", "Electronics", "Electronics", "Electronics", "Electronics", "Electronics"],
    "Sales": [60000, 20000, 30000, 65000, 22000, 28000],
    "Profit": [8000, 3000, 4000, 9000, 3500, 4200]
}

df = pd.DataFrame(data)
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
sales_by_product = df.groupby("Product")["Sales"].sum()
sales_by_product
```

```python
plt.figure(figsize=(6,4))
sns.barplot(x=sales_by_product.index, y=sales_by_product.values)
plt.title("Total Sales by Product")
plt.xlabel("Product")
plt.ylabel("Sales")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.barplot(x="Product", y="Profit", data=df)
plt.title("Profit by Product")
plt.xlabel("Product")
plt.ylabel("Profit")
plt.show()
```

```python
print("Key Insights:")
print("- Laptops generate the highest sales and profit.")
print("- Mobiles show moderate sales with consistent profit.")
print("- Tablets have stable but lower performance.")
```
